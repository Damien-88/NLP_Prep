### Key Takeaways

#### Markov Models (Chapter 9)
- Markov models represent sequences where the next state depends only on a fixed number of previous states (Markov assumption)
- Useful for language modeling and sequence prediction
- Hidden Markov Models (HMMs) extend Markov models with hidden states that generate observed symbols

- HMMs involve three fundamental problems:
    1) Probability of an observation sequence: solved using the forward algorithm
    2) Best state sequence: solved using the Viterbi algorithm
    3) Parameter estimation: solved using the Baum-Welch / EM algorithm

- Implementation considerations:
    - Efficient dynamic programming is critical for long sequences
    - Log probabilities prevent numerical underflow
    - Variants include higher-order Markov models and context-dependent emissions

- Properties:
    - Captures sequential dependencies probabilistically
    - Can model latent structures (e.g., POS tags as hidden states)

- Limitations:
    - Independence assumptions may oversimplify real language phenomena
    - Sparse data requires smoothing and careful initialization

#### Part-of-Speech Tagging (Chapter 10)

- POS tagging assigns syntactic categories to words in context

- Information sources:
    - Lexical features, neighboring tags, morphological clues
    - Previous tags are strong predictors in Markov models

- Markov model taggers:
    - Use transition probabilities 𝑃(𝑡𝑎𝑔<sub>𝑖</sub> ∣ 𝑡𝑎𝑔𝑖−<sub>1</sub>)
    - Emission probabilities 𝑃(𝑤𝑜𝑟𝑑<sub>𝑖</sub> ∣ 𝑡𝑎𝑔<sub>𝑖</sub>) model lexical distribution

- Applying HMMs to POS tagging:
    - Viterbi finds the most probable tag sequence
    - Forward-backward can compute marginal probabilities for each tag
    - Initialization affects convergence in EM-based training

- Alternative approaches:
    - Transformation-based learning (Brill tagging)
    - Automata-based models
    - Other statistical models (e.g., maximum entropy, CRFs)

- Cross-linguistic considerations:
    - Morphological complexity, contractions, clitics affect tagging
    - French: gender/number agreement, contractions (e.g., “du,” “au”), clitics (e.g., “l’homme”)
    - Russian and German introduce additional challenges (to be considered in later weeks)

- Evaluation:
    - Tagging accuracy is measured on held-out corpora
    - Taggers can inform parsing, information extraction, and downstream NLP tasks
