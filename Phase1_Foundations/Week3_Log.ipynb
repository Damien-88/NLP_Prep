### Key Takeaways

#### Context-Free Grammars (CFGs)
- CFGs model hierarchical syntactic structure using rewrite rules (e.g., S → NP VP)
- Unlike sequence models, CFGs explicitly encode constituency relationships between phrases
- They serve as a foundational formalism for constituency parsing tasks

- Key properties:
    - Capture hierarchical structure rather than linear dependencies
    - Define valid sentence structures through recursive rule application
    - Require simplification (CNF) for efficient parsing algorithms like CKY

- Chomsky Normal Form (CNF):
    - Restricts grammar rules to binary or lexical productions (A → BC or A → w)
    - Enables dynamic programming-based parsing
    - Introduces a trade-off between linguistic naturalness and computational efficiency

- Limitations:
    - Requires manually defined or learned grammars for coverage
    - Struggles with linguistic variation without extensive rule sets
    - Does not inherently model probabilities unless extended to PCFGs

#### Constituency Parsing and CKY Algorithm
- Constituency parsing aims to recover hierarchical phrase structure from sentences using CFGs
- CKY is a bottom-up dynamic programming algorithm for parsing sentences under CNF grammars

- Core mechanism:
    1) Initialize lexical entries on the diagonal of the parse table
    2) Combine spans iteratively using grammar rules
    3) Build larger constituents until the full sentence span is derived

- Key properties:
    - Time complexity: O(n³) for sentence length n
    - Guarantees completeness for CNF-compliant grammars
    - Efficiently explores all valid parse combinations within grammar constraints

- Implementation considerations:
    - Backpointers are required to reconstruct parse trees from the chart
    - Grammar sparsity directly limits parse coverage
    - Performance depends heavily on how well rules approximate real syntax

#### Cross-Linguistic Parsing Considerations (Russian Context)
- Russian presents challenges for constituency parsing due to:
    - Flexible word order, reducing reliance on fixed positional patterns
    - Rich morphological case marking that encodes syntactic roles
    - Strong dependency structure not always aligned with constituency assumptions

- Implications for CFG/CKY parsing:
    - Manually constructed grammars will have limited coverage
    - Many valid sentences may fail to parse due to rigid rule constraints
    - Morphological cues must be indirectly encoded in grammar rules to improve coverage

## Work Completed

### 1. Redesigned Russian CFG to better reflect real syntax
Significantly expanded original CKY grammar to better represent Russian sentence structure, moving away from an English-style SVO bias.

Added rules for:

- **Free word order**
  - S → NP VP
  - S → VP NP
  - S → VP

- **Copula-drop constructions**
  - S → NP NP (e.g. “он студент”)

- **Predicative expressions**
  - S → ADV ADJ
  - S → ADV NP

- **Coordination (CKY-safe binarization)**
  - S → S S_CONJ → CCONJ S

- **Expanded NP structure**
  - ADJ NOUN, NUM NOUN, NOUN NOUN
  - NP PP attachment
  - apposition

- **Expanded VP structure**
  - VERB NP, VERB PP
  - VP PP recursion
  - negation (PART VP, PART VERB)
  - adverb placement (ADV VERB / VERB ADV)

Original grammar was too English-like and structurally limited, so it could not capture common Russian constructions like copula-drop, flexible word order, and predicative sentences.

### 2. Introduced a UPOS-based CKY parsing mode
Added a second parser that operates on UPOS tag sequences instead of words.

**Key idea:**
- Replace each token with its UD POS tag
- No lexical lookup required
- No OOV (out-of-vocabulary) issues
  
This isolates structural grammar performance from lexical coverage, allowing evaluation of whether the CFG itself can generate valid Russian syntactic structures without vocabulary interference.

### 3. Rebuilt the dataset pipeline using UD properly
Rebuilt the evaluation pipeline around the Russian UD treebank:

- Loaded CoNLL-U format data
- Created a clean train/eval split:
  - 400 sentences for training
  - 50 sentences for evaluation (length-filtered)
- Removed punctuation consistently
- Ensured no overlap between training and evaluation data

Previous setup mixed training and evaluation data implicitly, which made results unreliable and inflated performance.

### 4. Improved lexical handling using UD + lemmatization
Rebuilt lexical rule extraction directly from UD data:

- Lexicon is now:
  - UPOS → word/lemma mappings
- Optional lemmatization using pymorphy3
- Words are normalized to lemmas when enabled

Used pymorphy3 (Russian morphological analyzer) to reduce inflectional sparsity.

Russian has heavy morphology (case, gender, number), so surface forms caused unnecessary sparsity. Lemmatization improves lexical generalization across forms.

### 5. Separated evaluation into two clear metrics
Introduced a two-layer evaluation system:

- **UPOS-sequence parsing (structure-only)**
  - Measures pure syntactic coverage of the grammar
  - No lexical influence

- **Word-level parsing (full system)**
  - Measures combined lexical + structural performance
  - Reflects real parsing ability on Russian sentences

## Observations after the changes

After running the updated system:

- UPOS structural success rate: **38% (19/50)**
- Word-level success rate: **4% (2/50)**
- Lexical rules added: **3169**
- Training sentences: **400**
- Evaluation sentences: **50**

## What this means

The results clearly show a split between structure and lexicon:

- The CFG has **moderate structural coverage** (38% UPOS success)
- The full system fails mainly due to **lexical and morphological mismatch**
- Word-level performance collapses despite lemmatization

This shows the grammar is partially effective, but vocabulary coverage and mapping between words and categories is the main bottleneck.

## Overall

I didn’t just improve the parser, I had to restructured it into a **two-layer evaluation framework**:

- One layer isolates **syntactic structure (UPOS mode)**
- One layer evaluates **full lexical + syntactic parsing (word mode)**

This makes the system’s failure modes measurable and separates grammar design issues from lexical/morphological issues.

## Key takeaway

The most important outcome today was not improved accuracy, but making it possible to clearly distinguish:

- structural grammar coverage
- lexical coverage
- morphological effects

which were previously conflated in the original version.
