### Key Takeaways

#### Context-Free Grammars (CFGs)
- CFGs model hierarchical syntactic structure using rewrite rules (e.g., S → NP VP)
- Unlike sequence models, CFGs explicitly encode constituency relationships between phrases
- They serve as a foundational formalism for constituency parsing tasks

- Key properties:
    - Capture hierarchical structure rather than linear dependencies
    - Define valid sentence structures through recursive rule application
    - Require simplification (CNF) for efficient parsing algorithms like CKY

- Chomsky Normal Form (CNF):
    - Restricts grammar rules to binary or lexical productions (A → BC or A → w)
    - Enables dynamic programming-based parsing
    - Introduces a trade-off between linguistic naturalness and computational efficiency

- Limitations:
    - Requires manually defined or learned grammars for coverage
    - Struggles with linguistic variation without extensive rule sets
    - Does not inherently model probabilities unless extended to PCFGs

#### Constituency Parsing and CKY Algorithm
- Constituency parsing aims to recover hierarchical phrase structure from sentences using CFGs
- CKY is a bottom-up dynamic programming algorithm for parsing sentences under CNF grammars

- Core mechanism:
    1) Initialize lexical entries on the diagonal of the parse table
    2) Combine spans iteratively using grammar rules
    3) Build larger constituents until the full sentence span is derived

- Key properties:
    - Time complexity: O(n³) for sentence length n
    - Guarantees completeness for CNF-compliant grammars
    - Efficiently explores all valid parse combinations within grammar constraints

- Implementation considerations:
    - Backpointers are required to reconstruct parse trees from the chart
    - Grammar sparsity directly limits parse coverage
    - Performance depends heavily on how well rules approximate real syntax

#### Cross-Linguistic Parsing Considerations (Russian Context)
- Russian presents challenges for constituency parsing due to:
    - Flexible word order, reducing reliance on fixed positional patterns
    - Rich morphological case marking that encodes syntactic roles
    - Strong dependency structure not always aligned with constituency assumptions

- Implications for CFG/CKY parsing:
    - Manually constructed grammars will have limited coverage
    - Many valid sentences may fail to parse due to rigid rule constraints
    - Morphological cues must be indirectly encoded in grammar rules to improve coverage

#### 1. CFG Grammar Construction

**Insights:**
- A minimal grammar is useful for validating parser logic before scaling to more realistic syntax
- Russian morphology and flexible constituent order make even simple grammar design nontrivial
- Grammar consistency matters immediately, since missing or mismatched categories prevent full parses

#### 2. CKY Parser Implementation
- Implemented a CKY parser using a chart of sets for constituent storage
- Separated grammar rules into lexical, unary, and binary rule groups
- Added unary-closure handling so categories like `NP -> PRON` and `NP -> NOUN` are propagated correctly
- Ensured the parser fills the diagonal with lexical categories and then combines spans bottom-up
- Added a helper check to determine whether the full sentence span is licensed by the start symbol `S`

**Insights:**
- Unary closure is necessary for this grammar; without it, valid toy sentences can fail to parse
- CKY parsing is straightforward structurally, but correctness depends on careful distinction between lexical, unary, and binary productions
- The chart-based design makes it easy to inspect intermediate constituents for debugging

#### 3. Toy Sentence Testing
- Tested the parser on the controlled sentence `я вижу собаку`
- Verified that the toy sentence receives a valid parse under the grammar
- Printed non-empty chart cells to inspect which constituents were assigned across spans

**Insights:**
- Small controlled examples are effective for validating parser behavior before moving to corpus data
- Chart inspection helps confirm whether failures are due to grammar coverage or algorithmic bugs

#### 4. UD Data Loading and Sample Evaluation
- Added a loader for Russian Universal Dependencies data in CoNLL-U format
- Extracted tokenized sentences while skipping comments, multiword tokens, and empty nodes
- Lowercased tokens and evaluated the parser on a small sample of Russian sentences
- Recorded parse outcomes for each sampled sentence
- Calculated a simple success rate over the evaluated sample

**Insights:**
- Real UD sentences quickly expose the limited coverage of a hand-built CFG
- Many parsing failures are linguistically meaningful, reflecting the mismatch between rigid constituency assumptions and Russian sentence variation
- Corpus-based evaluation is necessary even for small prototype grammars, because toy examples alone can hide coverage problems
