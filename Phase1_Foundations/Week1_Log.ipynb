### Week 1: German Language Models

- This week focuses on implementing and analyzing statistical language models using German text data, with the goal of building foundational intuition for probabilistic approaches to natural language processing. Specifically, unigram and bigram models are constructed and evaluated using corpora from Universal Dependencies, emphasizing how token distribution, word order, and morphological complexity influence model behavior.

- The choice to begin with German reflects its relevance as a morphologically rich, compounding language, which introduces challenges not present in simpler or more isolating languages. This aligns with the broader structure of the project, where each week pairs a core NLP task with a different language (German, French, Russian) in order to expose language-specific phenomena early. This design encourages comparative, cross-linguistic reasoning rather than overfitting intuition to English-centric assumptions.

- Through this process, the expected learning outcomes include: 

    1) A practical understanding of n-gram language modeling and probability estimation. 

    2) Familiarity with preprocessing and working with structured linguistic datasets. 

    3) Insight into how linguistic properties—such as compounding and vocabulary sparsity—affect statistical models. 

- Additionally, this week introduces reproducible experimentation practices, including structured logging, metric tracking (e.g., perplexity), and systematic observation.

- The expected outcome is a working implementation of unigram and bigram language models, accompanied by quantitative evaluation and qualitative analysis of their performance on German text. By the end of the week, the work should demonstrate both technical correctness and an initial level of research-oriented reflection, forming a foundation for more advanced neural and multilingual modeling in subsequent phases.

### Key Takeaways

#### Language Modeling (Chapter 3)
  - N-gram models approximate sequence probability using the Markov assumption
  - Data sparsity is a central challenge; higher-order models increase sparsity
  - Smoothing is necessary to avoid zero probabilities
  - Interpolation combines multiple n-gram orders:
    - Balances specificity (higher n) with robustness (lower n)
    - Important for real-world performance

#### Probabilistic & Linear Models (Chapter 4)

  ##### Logistic Regression
  - Models probability directly using a linear combination of features
  - Outputs are passed through a sigmoid function for binary classification
  - Can be extended to multiclass classification using softmax

  ##### Sigmoid & Softmax
  - Sigmoid maps real values to probabilities in [0,1]
  - Softmax generalizes this to multiple classes:
    - Produces a normalized probability distribution over labels
    - Sensitive to relative differences between logits

  #### Information Theory Concepts
  - Entropy measures uncertainty in a distribution
  - Cross-entropy measures how well a model predicts a true distribution
  - Minimizing cross-entropy is equivalent to maximizing likelihood

  #### Training & Optimization
  - Batch training processes multiple examples at once:
    - More stable gradient estimates than single-instance updates
    - Foundation for modern deep learning methods

  #### Evaluation Metrics

  ##### Confusion Matrix
  - Provides detailed breakdown of predictions vs actual labels
  - Useful for diagnosing specific error types

  ##### Precision, Recall, F1 Score
  - Precision: correctness of positive predictions
  - Recall: coverage of actual positives
  - F1 balances precision and recall

  ##### Macro vs Micro Averaging
  - Macro: treats all classes equally
  - Micro: weights by frequency
  - Important for imbalanced datasets

  #### Regularization
  - Prevents overfitting by penalizing large weights
  - Encourages simpler, more generalizable models
  - Critical when working with sparse or high-dimensional features

  ### Reflection
  - Even "simple" models (n-grams, logistic regression) rely heavily on:
    - Probability theory
    - Optimization
    - Careful evaluation
  - Many concepts (cross-entropy, softmax, regularization) directly carry into neural NLP models
